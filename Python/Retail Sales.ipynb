## Removing Missing Values

- Rows with missing CustomerID and Description are removed to ensure clean and reliable data.
- Records without CustomerID represent anonymous transactions, which cannot be used for customer-level analysis like segmentation, retention, or RFM. So I removed them to maintain data quality

```python
df = df.dropna(subset=['CustomerID'])
```

```python
df = df.dropna(subset=['Description'])
```

```python
df['CustomerID'] = df['CustomerID'].astype(int)
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

## We are creating a new column 'Revenue' which is calculated as:
 Revenue = Quantity Ã UnitPrice. This helps in analyzing sales performance.

```python
df['Revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
df['Description'] = df['Description'].str.strip()
df['Country'] = df['Country'].str.title()
```

```python
df.info()
```

```python
# Check null values
df['InvoiceDate'].isnull().sum()

# Fill missing dates (example)
#df['InvoiceDate'].fillna('2020-01-01', inplace=True)
df['InvoiceDate'] = df['InvoiceDate'].fillna('2020-01-01')

# OR drop rows with null dates
df = df[df['InvoiceDate'].notna()]
```

```python
# Convert InvoiceDate column to proper datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%d-%m-%Y %H:%M')
```
